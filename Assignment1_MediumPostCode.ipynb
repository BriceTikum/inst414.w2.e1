```python
import json
import pandas as pd
```

```python
act_Type = ["education", "recreational", "social", "diy", "charity", "cooking", "relaxation", "music", "busywork"]

activity_list = []

for activity in act_Type:
    boredActInfo = rq.get('http://www.boredapi.com/api/activity?type=' + activity).json()
    
    if 'error' not in boredActInfo:
        activity_list.append([boredActInfo['activity'], boredActInfo['type'], boredActInfo['participants'], boredActInfo['price'], boredActInfo['link'], boredActInfo['key'], boredActInfo['accessibility']])


dataset = pd.DataFrame(activity_list)
dataset.sample(9)
```
